```python
import csv
import numpy as np 
import os
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import normalize
import laplace as lap
```

```python
file_name = 'bank.txt'
csv_list = []
with open (file_name,newline='') as csvfile:
    bankreader = csv.reader(csvfile,delimiter = ',')
    for row in bankreader:
        csv_list.append(row)
```

```python
csv_array = np.array(csv_list[1:])
train_x = csv_array[:,1:13]
train_y_house = list(map(int,csv_array[:, -1]))
train_y_job = list(map(int,csv_array[:,-2]))

new_train_x = [] 
for i in train_x:
    q = []
    for j in i:
        q.append(int(j))
    new_train_x.append(q)

train_x_n = normalize(new_train_x, axis = 0, norm= 'max')
```

```python
classifier1 = KNeighborsClassifier(n_neighbors = 2, metric = 'minkowski', p =2)
classifier1.fit(train_x_n[0:-400],train_y_house[0:-400])
classifier1.fit(train_x_n[0:-400],train_y_job[0:-400])
```

```python
test_x = train_x_n[-400:]
test_y_house = train_y_house[-400:]
# y_pred_house = classifier1.predict(test_x)
# error = np.mean(test_y_house!=y_pred_house)
# print('The House accuracy is' + str(1-error))
print('The House accuracy is',end = ' : ')
print(classifier1.score(test_x,test_y_house))
```

```python
test_y_job = train_y_job[-400:]
# y_pred_job= classifier1.predict(test_x)
# error = np.mean(test_y_job!=y_pred_job)
# print('The Job accuracy is ' + str(1-error))
print('The Job accuracy is',end = ' : ')
print(classifier1.score(test_x,test_y_job))
```
